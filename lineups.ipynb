```python
import pandas as pd
import requests
```

```python
match_id = input("Enter the match ID: ")
url = f"https://www.sofascore.com/api/v1/event/{match_id}/lineups"

response = requests.get(url)
lineups = response.json()

home_players = lineups['home']['players']

for player in home_players:
    player['team'] = 'home'

away_players = lineups['away']['players']

for player in away_players:
    player['team'] = 'away'

all_players = home_players + away_players

df = pd.json_normalize(all_players)

df
##improvements to be made: -find a way to get it to recognize id by selecting a match, 
#-filter out irrelevant columsns and cleaning data, displaying home and away lineups when prompted in a visual
```
